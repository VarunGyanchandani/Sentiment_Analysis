```python
from pyabsa import available_checkpoints
checkpoint_map= available_checkpoints()
```

```python
from pyabsa import ATEPCCheckpointManager
```

```python
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(checkpoint ='english',auto_device=True)
```

```python
examples=df.content.values.tolist()
```

```python
inference_source = examples
atepc_result = aspect_extractor.extract_aspect(inference_source=inference_source, pred_sentiment=True)
```

```python
atepc_result
df = pd.DataFrame(atepc_result)
df
```

```python
df=df[df.astype(str)['aspect'] != '[]']
df
```

```python
from google.colab import files
df_final=df[['sentence','aspect','sentiment','confidence']]
df_final
files.download('df_final')
```

```python
special_char = "'[]"

# using join() + generator to remove special characters
df_final['aspect'] = [''.join(x for x in string if not x in special_char) for string in df['aspect'].astype(str)]
df_final['sentiment'] = [''.join(x for x in string if not x in special_char) for string in df['sentiment'].astype(str)]
df_final['confidence'] = [''.join(x for x in string if not x in special_char) for string in df['confidence'].astype(str)]
df_final.reset_index()
```

```python
pd.DataFrame(df_final).to_csv(brand + ' features.csv',index=False)
```
